```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = "https://www.nairaland.com/"
web_data = requests.get(url)
scraped_data = web_data.content
scraped_data
```

```python
nairaland = bs(scraped_data, 'lxml')
nairaland
```

```python
nairaland.findAll('tr')[9].text
```

```python
birthday_list = nairaland.findAll('tr')[9].text
```

```python
birthday_list = birthday_list.replace(',', '')
birthday_list = birthday_list.replace('Birthdays:', '')
```

```python
birthday_list = birthday_list.split(' ')
birthday_list
```

```python
nairaland_birthday_data = pd.DataFrame(birthday_list, columns = ["Username(age)"])
```

```python
nairaland_birthday_data
```

```python
#export data to a CSV
nairaland_birthday_data.to_csv("Nairaland_birthday_data.csv", encoding = 'utf-8')
```
